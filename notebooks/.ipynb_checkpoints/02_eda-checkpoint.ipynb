```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
plt.rcParams["figure.figsize"] = (8, 5)

data_path = "../data/processed/diabetic_data_clean.csv"
df = pd.read_csv(data_path)

df.shape, df.head()
```

```python
df["readmitted_30"].value_counts()
df["readmitted_30"].value_counts(normalize=True)
```

```python
sns.countplot(x="age", hue="readmitted_30", data=df)
plt.xticks(rotation=45)
plt.title("Readmission by Age Group")
plt.show()
```

```python
sns.boxplot(x="readmitted_30", y="time_in_hospital", data=df)
plt.title("Time in Hospital vs Readmission")
plt.show()
```

```python
sns.countplot(x="change", hue="readmitted_30", data=df)
plt.title("Medication Change vs Readmission")
plt.show()
```

```python
sns.countplot(x="insulin", hue="readmitted_30", data=df)
plt.title("Insulin Use vs Readmission")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"])
corr = numeric_cols.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
